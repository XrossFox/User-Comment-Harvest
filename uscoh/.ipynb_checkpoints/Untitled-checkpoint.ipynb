```python
from stackapi import StackAPI

SITE = StackAPI('stackoverflow')
SITE.page_size = 1
SITE.max_pages = 2
response = SITE.fetch('search', intext='haskell', tagged='haskell', sort="relevance",
                      filter="!7qBwspMQR3L7c4q7tesaRX(_gP(rj*U-.H")
#for key in comments:
    #print("{} - {}".format(key,type(comments[key])))
print("Devuelve una respuesta: {}".format(type(response)))
print('Y contiene:')

for keys in response:
    print('-'*4+" {} - que es: {}".format(keys,type(response[keys])))
    
    if keys == 'items':
        items = response[keys]
        print("-"*8+" Esta lista contiene preguntas:")
        
        for question in items:
            print("-"*8+" pregunta: {} - {}".format(type(question),repr(question)[:30]))
            for q_key in question:
                print("-"*12+" {} que es: {}".format(q_key,type(question[q_key])))
                
                if q_key == 'comments':
                    for comment in question[q_key]:
                        print("-"*16+" elemento: {}".format(type(comment)))
                        
                        for c_item in comment:
                            print("-"*20+" {} que es: {}".format(c_item,type(c_item)))
                
                if q_key == 'answers':
                    for comment in question[q_key]:
                        print("-"*16+" elemento: {}".format(type(comment)))
                        for c_item in comment:
                            print("-"*20+" {} que es: {}".format(c_item,type(c_item)))
                
                
                            
    

#question = comments['items']
#for keys in question[0]:
 #   print("{} - {} - {}".format(keys, type(question[0][keys]), str(question[0][keys])[:60]))
    
#print("-"*10)

#all_comments = []
#question_comments = question[0]["comments"]
#print(type(question_comments))

#for comm in question_comments:
    #print("{} - {}".format(comm,str(question_comments[comm])[:60]))
 #   print(comm)

```
